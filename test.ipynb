```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True

df2[~df2['total_sqft'].apply(is_float)].head(10)
```

```python
def convert_sqft_to_num(x):
    if isinstance(x, str):
        # Handle ranges like "1000-1200"
        tokens = x.split('-')
        if len(tokens) == 2:
            return (float(tokens[0]) + float(tokens[1])) / 2
        try:
            return float(x)  # Handle single numbers like "1200"
        except:
            return None
    # If it's already a number (or NaN), return as is
    try:
        return float(x)
    except:
        return None
```

```python
df3 = df2.copy()
df3['total_sqft_num'] = df3['total_sqft'].apply(convert_sqft_to_num) 
df3['total_sqft_num'].unique()
```
